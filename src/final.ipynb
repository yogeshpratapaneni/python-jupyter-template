```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 11)  # x values from 0 to 10
y = x**2  # y values as the square of x

# Plot
plt.plot(x, y, marker='o')

# Customizations
plt.title('Line Plot of y = x^2')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)

# Show plot
plt.show()
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['blue', 'orange', 'green', 'red']  # Different colors for each bar

# Plot
plt.bar(categories, values, color=colors)

# Customizations
plt.title('Bar Plot of Categories and Values')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show plot
plt.show()
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plot
plt.scatter(x, y)

# Annotations
for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Customizations
plt.title('Scatter Plot with Annotations')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)

# Show plot
plt.show()
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']  # Optional colors for each slice

# Plot
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

# Customizations
plt.title('Pie Chart of Programming Languages')
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.

# Show plot
plt.show()
import matplotlib.pyplot as plt

# Data for the plots
x_line = range(1, 11)
y_line = [i**2 for i in x_line]

categories_bar = ['A', 'B', 'C', 'D']
values_bar = [5, 7, 3, 9]

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]

# Create 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line plot in the first subplot
axs[0, 0].plot(x_line, y_line, marker='o')
axs[0, 0].set_title('Line Plot')
axs[0, 0].set_xlabel('X Values')
axs[0, 0].set_ylabel('Y Values')

# Bar plot in the second subplot
axs[0, 1].bar(categories_bar, values_bar, color=['blue', 'orange', 'green', 'red'])
axs[0, 1].set_title('Bar Plot')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter plot in the third subplot
axs[1, 0].scatter(x_scatter, y_scatter)
for i in range(len(x_scatter)):
    axs[1, 0].annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(0, 10), ha='center')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X Values')
axs[1, 0].set_ylabel('Y Values')

# Pie chart in the fourth subplot
axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', startangle=140, colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'])
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
import matplotlib.pyplot as plt

# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Plot
plt.hist(data, bins=5, color='skyblue', edgecolor='black')  # Customize the color and edge color of the bars

# Customizations
plt.title('Histogram of Data')
plt.xlabel('Data Values')
plt.ylabel('Frequency')

# Show plot
plt.show()
```
